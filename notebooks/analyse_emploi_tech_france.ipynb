## 5. 💡 Insights & Conclusions

### Ce que révèle cette analyse :

```python
# --- Résumé des insights ---

top_comp = df_comp['competence'].value_counts().head(3)
top_ville = df['ville'].value_counts().idxmax()
pct_paris = df['ville'].value_counts(normalize=True)['Paris'] * 100
pct_stage = df['type_contrat'].value_counts(normalize=True)['Stage'] * 100
salaire_stage_median = df[df['type_contrat'] == 'Stage']['salaire'].median()

print('=' * 55)
print('       📊 INSIGHTS CLÉS DU MARCHÉ TECH & IA FR')
print('=' * 55)
print(f'  🥇 Compétence n°1 : {top_comp.index[0]} ({top_comp.iloc[0]} offres)')
print(f'  🥈 Compétence n°2 : {top_comp.index[1]} ({top_comp.iloc[1]} offres)')
print(f'  🥉 Compétence n°3 : {top_comp.index[2]} ({top_comp.iloc[2]} offres)')
print(f'  🏙️  Ville dominante : {top_ville} ({pct_paris:.0f}% des offres)')
print(f'  📄 Part des stages : {pct_stage:.1f}% des offres')
print(f'  💶 Salaire médian stage : {salaire_stage_median:.0f} €/mois')
print('=' * 55)
```
